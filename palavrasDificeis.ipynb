#enunciado 2 ✅
Em um sistema de estacionamento, o valor cobrado depende do número
de horas que o carro ficou estacionado. A cada hora completa, é cobrado R$ 5,00, e o valor
total é calculado somente se o tempo for maior que zero. O código a seguir verifica o tempo
estacionado e retorna o valor a ser pago:
function calculaEstacionamento(horasTotais) {
if (horasTotais > 0) {
let valorTotal = horasTotais * 5;
return `Total a pagar: R$ ${valorTotal}`;
} else {
return "Nenhum valor a pagar.";
}
}

**.else
.return**
#significado dessas palavras ✅
 A função calculaEstacionamento usa uma estrutura condicional if...else para tomar uma decisão:

if (horasTotais > 0): O código verifica a condição. Se o tempo total for maior que zero (ou seja, o carro ficou estacionado por algum tempo válido), o código dentro do if é executado, calcula o valor e retorna o total a pagar.

else: Esta palavra-chave significa "caso contrário". Se a condição do if (tempo maior que zero) for falsa (ou seja, o tempo é zero ou negativo), o código dentro do bloco else é executado.

O Papel do return dentro do else
JavaScript
